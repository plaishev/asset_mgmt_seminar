```python
import pandas as pd
import matplotlib.pyplot as plt

# Load each portfolio's CSV into a separate DataFrame
one_over_N_df = pd.read_csv("./portfolio_csv/one_o_N_portfolio3.csv", index_col=0, parse_dates=True)
sixtyE_fourtyB_df = pd.read_csv("./portfolio_csv/sixtyE_fourtyB_portfolio3.csv", index_col=0, parse_dates=True)
sixtyE_fourtyG_df = pd.read_csv("./portfolio_csv/sixtyE_fourtyG_portfolio3.csv", index_col=0, parse_dates=True)
min_var_df = pd.read_csv("./portfolio_csv/min_var_portfolio3.csv", index_col=0, parse_dates=True)
max_div_df = pd.read_csv("./portfolio_csv/max_div_portfolio3.csv", index_col=0, parse_dates=True)
risk_parity_df = pd.read_csv("./portfolio_csv/risk_parity_portfolio3.csv", index_col=0, parse_dates=True)
max_sharpe_df = pd.read_csv("./portfolio_csv/max_sharpe_portfolio3.csv", index_col=0, parse_dates=True)

# Plot them all in one figure
plt.figure(figsize=(10, 6))
#plt.plot(one_over_N_df.index, one_over_N_df["Portfolio"], label="1/N")
#plt.plot(sixtyE_fourtyG_df.index, sixtyE_fourtyB_df["Portfolio"], label="60/40B")
#plt.plot(sixtyE_fourtyG_df.index,sixtyE_fourtyG_df["Portfolio"], label="60/40G")
plt.plot(min_var_df.index, min_var_df["Portfolio"], label="Min Var")
plt.plot(max_div_df.index, max_div_df["Portfolio"], label="Max Div")
plt.plot(risk_parity_df.index, risk_parity_df["Portfolio"], label="Risk Parity")
plt.plot(max_sharpe_df.index, max_sharpe_df["Portfolio"], label="Max Sharpe")

plt.title("Covariance Based Portfolios Summary")
plt.xlabel("Date")
plt.ylabel("Value")
plt.ylim(0, 5000000)
plt.legend()
plt.grid(True, linestyle='--', alpha=0.6)
plt.show()
```
